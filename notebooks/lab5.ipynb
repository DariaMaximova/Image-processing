# Лабораторная работа 5. Цифровая обработка изображений

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# load image and do HSV
image = cv2.imread('/content/lab56.jpg')
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Range fpr blue mask
lower_blue= np.array([100, 50, 143])
upper_blue = np.array([130, 255, 255])

# Create mask
mask1 = cv2.inRange(hsv_image, lower_blue, upper_blue)


# no noise for mask
kernel = np.ones((5, 5), np.uint8)
mask = cv2.morphologyEx(mask1, cv2.MORPH_OPEN, kernel)

# Searching countour
contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
result = image

# result
plt.figure(figsize=(15,20))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(hsv_image, cv2.COLOR_HSV2RGB))
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(mask, cv2.COLOR_GRAY2BGR))
plt.show()
```
