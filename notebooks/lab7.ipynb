# Лабораторная работа 7. Цифровая обработка изображений

## Задание: Найти шаблон

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Загрузка ищображений
image = cv2.imread('/content/image7.jpg')
pattern_image = cv2.imread('/content/pattern7.jpg')
pattern_image = cv2.pyrDown(pattern_image)
image = cv2.pyrDown(image)


# Нахождение шаблона на изображении
result = cv2.matchTemplate(image, pattern_image, cv2.TM_CCOEFF_NORMED)

# Установка порога для определения совпадений
threshold = 0.73
locations = np.where(result >= threshold)
locations = list(zip(*locations[::-1]))

# Выделяем объекты
for loc in locations:
    cv2.rectangle(image, loc, (loc[0] + pattern_image.shape[1], loc[1] + pattern_image.shape[0]), (0, 0, 0), 3)

# Вывод изображений
cv2_imshow(pattern_image)
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
